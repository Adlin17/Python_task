```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)  # Data for the first y-axis
y2 = np.cos(x)  # Data for the second y-axis
y3 = np.tan(x)  # Data for the third y-axis

# Create a figure and primary axis
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot on the first y-axis
ax1.plot(x, y1, label="sin(x)", color="blue")
ax1.set_xlabel("X-Axis")
ax1.set_ylabel("sin(x)", color="blue")
ax1.tick_params(axis='y', labelcolor="blue")

# Create a second y-axis sharing the same x-axis
ax2 = ax1.twinx()
ax2.plot(x, y2, label="cos(x)", color="red")
ax2.set_ylabel("cos(x)", color="red")
ax2.tick_params(axis='y', labelcolor="red")

# Create a third y-axis sharing the same x-axis
ax3 = ax1.twinx()
ax3.spines["right"].set_position(("outward", 60))  # Offset the third y-axis
ax3.plot(x, y3, label="tan(x)", color="green")
ax3.set_ylabel("tan(x)", color="green")
ax3.tick_params(axis='y', labelcolor="green")

# Add a title
plt.title("Multiple Y-Axes Sharing a Common X-Axis")

# Add a legend (manually manage it for clarity)
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
lines_3, labels_3 = ax3.get_legend_handles_labels()
plt.legend(lines_1 + lines_2 + lines_3, labels_1 + labels_2 + labels_3, loc="upper left")

# Show the plot
plt.tight_layout()
plt.show()
```
